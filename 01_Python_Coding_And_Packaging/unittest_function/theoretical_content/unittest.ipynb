### **üß™ Tipos de Asser√ß√µes (`asserts`) no Unittest üß™**

*_Dentro do `unittest`, os m√©todos de asser√ß√£o s√£o a ferramenta principal para verificar se o seu c√≥digo se comporta como o esperado. Eles comparam um resultado obtido com um resultado esperado e falham o teste se a compara√ß√£o n√£o for verdadeira. Abaixo est√£o os mais comuns:_*

- **`assertEqual(a, b)`**
  - *_Verifica se `a` √© igual a `b`._*
  - *_Uso: O mais comum, para checar se uma fun√ß√£o retorna o valor exato esperado._*

- **`assertNotEqual(a, b)`**
  - *_Verifica se `a` √© diferente de `b`._*
  - *_Uso: √ötil para garantir que um valor n√£o seja algo espec√≠fico._*

- **`assertTrue(x)`**
  - *_Verifica se `x` √© `True`._*
  - *_Uso: Para checar condi√ß√µes booleanas, como "o item existe na lista?"._*

- **`assertFalse(x)`**
  - *_Verifica se `x` √© `False`._*
  - *_Uso: O oposto de `assertTrue`, para garantir que uma condi√ß√£o seja falsa._*

- **`assertIsNone(x)`**
  - *_Verifica se `x` √© `None`._*
  - *_Uso: Comum para checar fun√ß√µes que podem retornar `None` em caso de erro ou aus√™ncia de valor._*

- **`assertIn(a, b)`**
  - *_Verifica se `a` est√° contido em `b`._*
  - *_Uso: Para checar a presen√ßa de um item em uma cole√ß√£o._*

- **`assertIsInstance(a, b)`**
  - *_Verifica se `a` √© uma inst√¢ncia da classe `b`._*
  - *_Uso: Para garantir que sua fun√ß√£o est√° retornando um objeto do tipo correto._*

- **`assertRaises(excecao, funcao, *args, **kwargs)`**
  - *_Verifica se a chamada de `funcao(*args, **kwargs)` levanta a `excecao` esperada._*
  - *_Uso: Essencial para testar o tratamento de erros e exce√ß√µes no seu c√≥digo._*

#### **_üî© ESTRUTURA DO UNITTEST üî©_**

*_Para que o `unittest` funcione, voc√™ precisa seguir uma estrutura m√≠nima. O `unittest` descobre e executa os testes automaticamente se voc√™ seguir estas conven√ß√µes:_*

1.  ***Importar a biblioteca `unittest`***:
    -   *_Sempre comece com `import unittest`._*

2.  ***Criar uma Classe de Teste***:
    -   *_Seus testes devem estar dentro de uma classe que herda de `unittest.TestCase`._*
    -   *_Exemplo: `class MeuTeste(unittest.TestCase):`_*

3.  ***Criar M√©todos de Teste***:
    -   *_Cada teste individual √© um m√©todo dentro da sua classe._*
    -   ***‚ö†Ô∏è IMPORTANTE: O nome de cada m√©todo de teste DEVE come√ßar com o prefixo `test_`.*** *_√â assim que o `unittest` sabe que aquele m√©todo √© um teste que deve ser executado. Se n√£o come√ßar com `test_`, ele ser√° ignorado._*
    -   *_Exemplo: `def test_soma_positiva(self):`_*

4.  ***Usar Asser√ß√µes***:
    -   *_Dentro de cada m√©todo de teste, use um dos m√©todos `self.assert...()` (como `self.assertEqual()`) para verificar o resultado._*

5.  ***Ponto de Entrada para Execu√ß√£o***:
    -   *_Adicione o bloco `if __name__ == '__main__': unittest.main()` no final do arquivo. Isso permite que voc√™ execute os testes diretamente rodando o arquivo Python._*
